# Fortran Testing

The output file has 100 time steps, each with 100 points. The first data item is the step number. Values range from 0 to 1 along the **y** axis, and 0 to 100 along the **x** axis. 

We start off by loading the data.

```python
with open('output.txt','r') as fin:
    lines = fin.readlines()
    data = [line.rstrip().split() for line in lines]
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
```

```python
x = np.linspace(0,100,100)
%matplotlib notebook

fig, ax = plt.subplots()
line, = ax.plot([])     # A tuple unpacking to unpack the only plot
ax.set_xlim(0,100)
ax.set_ylim(0, 1)
def animate(i):
    yp = data[i][1:]
    y = []
    for j in yp:
        y.append(float(j))
    line.set_data((x, y))
    return line
    
anim = FuncAnimation(fig, animate, frames=100, interval=1)
plt.show()
```
